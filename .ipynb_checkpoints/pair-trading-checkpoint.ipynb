```python
"""Utility Functions"""
import random
def draw_slope_intercept_changes(prices, state_means):
    """
    Plot the slope and intercept changes from the 
    Kalman Filter calculated values.
    """
    pad.DataFrame(
        dict(
            slope=state_means[:, 0], 
            intercept=state_means[:, 1]
        ), index=prices.index
    ).plot(subplots=True)
    plt.title("Slope and Intercept From Dynamic Beta")
    plt.show()
    
def zscore(stocks):
    """ Basic function to compute z-score. """
    return (stocks - stocks.mean()) / np.std(stocks)


def create_sig(zscore_roll):
    """Create signal matrix for backtest"""
    sig = zscore_roll.copy()
    sig = sig.to_frame()
    sig.columns = ['z_score']
    upper =  0.5999999999999996
    lower = -0.95
#     upper = 1.0
#     lower = -1.0
    sig['z_score'].loc[(sig['z_score'].isnull())] = 0
    sig['z_score'].loc[(sig['z_score'].between(lower, upper))] = 0
    sig['z_score'].loc[(sig['z_score'] < lower)] = -1
    sig['z_score'].loc[(sig['z_score'] > upper)] = 1
    return sig, upper, lower

def create_rand_sig(zscore_roll):
    """Create signal matrix with random entry/exit parameters for backtest"""
    sig = zscore_roll.copy()
    sig = sig.to_frame()
    sig.columns = ['z_score']
    upper = random.uniform(0.0, 1.0)
    # upper = 0.9543606834263253
    lower = random.uniform(-1.0, 0.0)
    sig['z_score'].loc[(sig['z_score'].isnull())] = 0
    sig['z_score'].loc[(sig['z_score'].between(lower, upper))] = 0
    sig['z_score'].loc[(sig['z_score'] < lower)] = -1
    sig['z_score'].loc[(sig['z_score'] > upper)] = 1
    return sig, upper, lower

def create_zscore_roll(spread):
    """Create rolling z-score of spread"""
    spread_mavg1 = spread.rolling(5).mean()
    # 60 day rolling
    spread_mavg2 = spread.rolling(60).mean()
    spread_std = spread.rolling(60).std()
    # 20 day rolling
#     spread_mavg2 = spread.rolling(20).mean()
#     spread_std = spread.rolling(20).std()
    zscore_roll = (spread_mavg1 - spread_mavg2)/spread_std
    return zscore_roll
```

```python
""" Get dynamic hedge ratio. """
from pykalman import KalmanFilter
def calc_kalman(stock1, stock2):
    """
    Utilise the Kalman Filter from the pyKalman package
    to calculate the slope and intercept of the regressed
    ETF prices.
    """
    delta = 1e-5
    trans_cov = delta / (1 - delta) * np.eye(2)
    obs_mat = np.vstack(
        [stock1, np.ones(stock1.shape)]
    ).T[:, np.newaxis]
    
    kf = KalmanFilter(
        n_dim_obs=1, 
        n_dim_state=2,
        initial_state_mean=np.zeros(2),
        initial_state_covariance=np.ones((2, 2)),
        transition_matrices=np.eye(2),
        observation_matrices=obs_mat,
        observation_covariance=1.0,
        transition_covariance=trans_cov
    )
    state_means, state_covs = kf.filter(stock2.values)
    return state_means, state_covs
```

```python
"""Utility functions."""
from statsmodels.tsa.stattools import adfuller
def get_beta_hr(stockA, stockB):
    # Calculate optimal hedge ratio "beta" for linear combination
    
    # Log prices
    y = np.log(stockA).tolist()
    x = np.log(stockB).tolist()

    # Discrete prices
#     y = stockA.tolist()
#     x = stockB.tolist()
    x = sm.add_constant(x)
    
    # Perform regression to find ratio
    res = sm.OLS(y, x).fit()
    
    beta_hr = res.params[1]
    
    return beta_hr

def get_beta_dy(stock1, stock2):
    """Return dynamic beta hr of a pair."""
    beta_means, beta_covs = calc_kalman(stock1, stock2)
    betas = []
    for beta in beta_means:
        betas.append(beta[0])
    betas = pad.Series(betas)
    betas.index = stock2.index
    return betas, beta_means

def dickey_fuller(stockA, stockB):
    """ Cointegrated Augmented Dickey-Fuller Test. """
    
    # Get beta hedge ratio
    # beta_hr = get_beta_hr(stockA, stockB)
    betas, beta_means = get_beta_dy(np.log(stockA), np.log(stockB))
    
    # Calculate residuals of the linear combination
    spread = np.log(stockA) - betas*np.log(stockB)
    # spread = stockA - betas * stockB
   
    # Caculate and output ADF test on the residuals
    # If p value is under 0.05 we can reject null (they're cointegrated!)
    adf = adfuller(spread)
    
    # Return t-stat, p-value, crit
    return adf[0], adf[1], adf[4]

# NOTE: Subject to multiple comparison bias
def find_cointegrated_pairs(data):
    """Find co-integrated pairs."""
    n = data.shape[1]
    score_matrix = np.zeros((n , n))
    pvalue_matrix = np.ones((n, n))
    keys = data.keys()
    pairs = []
    for i in range(n):
        for j in range(i+1, n):
            S1 = data[keys[i]]
            S2 = data[keys[j]]
            result = coint(S1, S2)
            score = result[0]
            pvalue = result[1]
            score_matrix[i, j] = score
            pvalue_matrix[i, j] = pvalue
            if pvalue < 0.01:
                pairs.append((keys[i], keys[j], pvalue))
    return score_matrix, pvalue_matrix, pairs
```

```python
"""Run ADF Testing"""
def run_adf(prices, stock_pairs):
    pairs = []
    for pair in stock_pairs:
        result = dickey_fuller(prices[pair[0]], prices[pair[1]])
        score = result[0]
        pvalue = result[1]
        if pvalue < 0.01:
            pairs.append((pair[0], pair[1], pvalue))
    return pairs
```

```python
"""Convert prices dataframe to ndarray."""
def convert_df(df):
    X = []
    for column in df.columns:
        X.append(df[column].to_numpy())
    return np.array(X)

"""Get ticks belong to each cluster."""
def get_ticks(ticks, clusters):
    cluster_ticks = {}
    for i in range(len(ticks)):
        cluster = clusters[i]
        if cluster not in cluster_ticks:
            cluster_ticks[cluster] = []
        cluster_ticks[cluster].append(ticks[i])
    return cluster_ticks
```

```python
def create_specified_sig(zscore_roll, upper, lower):
    """Create signal matrix with specified parameters for backtest"""
    sig = zscore_roll.copy()
    sig = sig.to_frame()
    sig.columns = ['z_score']
    sig['z_score'].loc[(sig['z_score'].isnull())] = 0
    sig['z_score'].loc[(sig['z_score'].between(lower, upper))] = 0
    sig['z_score'].loc[(sig['z_score'] < lower)] = -1
    sig['z_score'].loc[(sig['z_score'] > upper)] = 1
    return sig


def fitness_backtest(upper, lower, stock1, stock2, bt_data, display):
    betas, beta_means = get_beta_dy(np.log(bt_data[stock1]), np.log(bt_data[stock2]))
    spread = np.log(bt_data[stock1]) - betas * np.log(bt_data[stock2])
    """"""
    # "Unfold" spread
    spread_broken = {'asset1': np.log(bt_data[stock1]), 'asset2': betas * np.log(bt_data[stock2]) * -1}
    unfold = pad.DataFrame(data = spread_broken)
    unfold.index = spread.index
    """"""
     # Plot z-score of spread
    if display:
        zscore(spread).plot()
        plt.title('Z-Score of Spread')
        plt.axhline(zscore(spread).mean(), c = 'b')
        plt.axhline(1.0, c = 'r', ls = '--')
        plt.axhline(-1.0, c = 'r', ls = '--')

    # Plot slope and intercept changes from Kalman Filter values
    zscore_roll = create_zscore_roll(spread)
    if display:
        draw_slope_intercept_changes(bt_data[stock1], beta_means)
        zscore_roll.plot(label = 'Rolling z score')
        plt.title("Rolling Z-Score of Spread")
        plt.axhline(0, color='black')
        plt.axhline(1, c='r',ls='--')
        plt.axhline(-1, c='r',ls='--')
    # Create signal matrix
    sig = create_specified_sig(zscore_roll, upper, lower)
    # Create random signal
    # sig, upper, lower = create_rand_sig(zscore_roll)
    bt_weights = copy.deepcopy(bt_data)
    bt_weights[stock1] = -sig
    bt_weights[stock2] = sig
    
    # Create strategy
    pair_trade = bt.Strategy('Pairs Trading ', [bt.algos.WeighTarget(bt_weights),
                                    bt.algos.Rebalance(),
                                    bt.algos.RunAfterDays(100)
                                    ])
    benchmark = bt.Strategy('Benchmark', [bt.algos.SelectRandomly(), 
                                          bt.algos.WeighEqually(),
                                          bt.algos.Rebalance(),
                                          bt.algos.RunAfterDays(100)
                                    ])
    # Run Backtest, compare to benchmark
    t = bt.Backtest(pair_trade, bt_data)
    b = bt.Backtest(benchmark, bt_data)
    res = bt.run(t,b)
    if display:
        res.display()
        res.plot()
    strat_sharpe = res[0].stats['monthly_sharpe']
    bench_sharpe = res[1].stats['monthly_sharpe']
    return strat_sharpe
```

```python
"""Run Backtest on Selected Pair."""
def backtest(pairs, data, display):
    weights = copy.deepcopy(data)
    # Divide by p to weigh equally
    p = 2*len(pairs)
    for pair in pairs:
        betas, beta_means = get_beta_dy(np.log(data[pair[0]]), np.log(data[pair[1]]))
        spread = np.log(data[pair[0]]) - betas * np.log(data[pair[1]])
         # Plot z-score of spread
        if display:
            zscore(spread).plot()
            plt.title('Z-Score of Spread')
            plt.axhline(zscore(spread).mean(), c = 'b')
            plt.axhline(1.0, c = 'r', ls = '--')
            plt.axhline(-1.0, c = 'r', ls = '--')

        # Plot slope and intercept changes from Kalman Filter values
        zscore_roll = create_zscore_roll(spread)
        if display:
            draw_slope_intercept_changes(data[pair[0]], beta_means)
            zscore_roll.plot(label = 'Rolling z score')
            plt.title("Rolling Z-Score of Spread")
            plt.axhline(0, color='black')
            plt.axhline(1, c='r',ls='--')
            plt.axhline(-1, c='r',ls='--')
        # Create signal matrix
        # sig, upper, lower = create_sig(zscore_roll)
        sig = create_specified_sig(zscore_roll, pair[2], pair[3])
        # Create random signal
        # sig, upper, lower = create_rand_sig(zscore_roll)
        weights[pair[0]] = -sig/p
        weights[pair[1]] = sig/p
        # Set Short Signals Based on spread
            #     bt_weights[stock1].loc[(bt_weights[stock1] == -1.0)] *= spread
            #     bt_weights[stock2].loc[(bt_weights[stock2] == -1.0)] *= spread
    # Create strategy
    pair_trade = bt.Strategy('Pairs Trading ', [bt.algos.WeighTarget(weights),
                                    bt.algos.Rebalance(),
                                    bt.algos.RunAfterDays(100),
                                    ])
    benchmark = bt.Strategy('Benchmark', [bt.algos.SelectRandomly(), 
                                          bt.algos.WeighEqually(),
                                          bt.algos.Rebalance(),
                                          bt.algos.RunAfterDays(100)
                                    ])
    # Run Backtest, compare to benchmark

    t = bt.Backtest(pair_trade, data)
    b = bt.Backtest(benchmark, data)
    res = bt.run(t,b)
#     if display:
#         res.display()
#         res.plot()
    res.display()
    res.plot()
    strat_sharpe = res[0].stats['monthly_sharpe']
    bench_sharpe = res[1].stats['monthly_sharpe']
    return strat_sharpe
#     if strat_sharpe > bench_sharpe:
#         return True, strat_sharpe, spread.pct_change(1).shift(-1)
#     return False, strat_sharpe, spread.pct_change(1).shift(-1)
```

```python
"""Backtest sequence of pairs and return best results."""
def calc_backtests(pairs, stocks):
    # Use monthly sharpe as primary metric
    data = []
    for pair in pairs:
        stock1 = pair[0]
        stock2 = pair[1]
        bt_data = stocks[[stock1, stock2]]
        res, sharpe = backtest(stock1, stock2, bt_data, False)
        if res:
            data.append((stock1, stock2, sharpe))
    data.sort(key = lambda x: x[2], reverse=True)
    return data
```
