```python
# Choose dataset to use
chosen_dataset = 'finance'
path = 'prices/{}.csv'.format(chosen_dataset)
```

```python
data = pd.read_csv(path, index_col=0)
data.index = pd.to_datetime(data.index)
data.head()
```

```python
import time
import datetime

stocks = data

# Split into train and test sets
split_date ='2019-01-01'
s = datetime.datetime.strptime(split_date,"%Y-%m-%d")
train = stocks.loc[stocks.index < s]
test = stocks.loc[stocks.index >= s]
```

```python
"""Retrieve Optimal Pairs"""
import json
path = "optimized_pairs/{}_pairs.json".format(chosen_dataset)
with open(path) as f:
    pairs_data = json.load(f)[chosen_dataset]
```

```python
def process_pairs_data(pairs_data, test_data):
    data = []
    test = []
    for pair in pairs_data:
        data.append([pair['stock1'], pair['stock2'], pair['upper'], pair['lower']])
        test.extend([pair['stock1'], pair['stock2']])
    return data, test_data[test]
```

```python
processed_pairs, backtest_data = process_pairs_data(pairs_data, test)
```
